```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
print("Basic Information:")
print(df.info())
```

```python
print("Summary Statistics")
print(df.describe())
```

```python
#Average rent of all
avg_rent=df['Rent'].mean()
print("Average rent in Rs:")
avg_rent
```

```python
#Average rent by locations
avg_rentby_location=df.groupby('Location')['Rent'].mean()
avg_rentby_location.sort_values()
```

```python
#BAR CHART
plt.figure(figsize=(10,6))
avg_rentby_location.plot(kind='bar',color='skyblue')
plt.title('Average Rent by Location')
plt.xlabel('Location')
plt.ylabel('Average Rent(INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Scatter Plot: Size Vs Rent
plt.figure(figsize=(10,6))
sns.scatterplot(x='Size',y='Rent',hue='Location',data=df,palette='Set2',s=10,alpha=0.7)
plt.title('Apartment Size vs Rent')
plt.xlabel('Size (sq.ft)')
plt.ylabel('Rent (INR)')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#HeatMap: Correlation matrix
plt.figure(figsize=(6,5))
corr_matrix=df[['Size','Bedroom','Rent']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
